```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('/work/final/final_final_combined_data_drop_submission.csv')

# Set the figure size
plt.figure(figsize=(12, 6))

# Create box plot with dot overlay using seaborn
# stripplot adds the individual points
# boxplot adds the box plot
sns.boxplot(x='Name_Category', 
            y='top5_accept_rating_criteria_mean',
            data=df,
            color='lightgray')

sns.stripplot(x='Name_Category',
             y='top5_accept_rating_criteria_mean',
             data=df,
             color='navy',
             alpha=0.3,
             size=4,
             jitter=0.2)

# Customize the plot
plt.title('Distribution of Top 5 Accept Rating Criteria by Name Category', 
          fontsize=14, 
          pad=15)
plt.xlabel('Name Category', fontsize=12)
plt.ylabel('Top 5 Accept Rating Criteria (Mean)', fontsize=12)

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()
```

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('/work/final/final_final_combined_data_drop_submission.csv')

# Set up the figure and grid
plt.figure(figsize=(12, 6))

# Create distribution plot using seaborn
sns.displot(data=df,
           x='top5_accept_rating_criteria_mean',
           hue='Name_Category',
           kind='kde',  # Kernel Density Estimation plot
           fill=True,   # Fill the density curves
           alpha=0.4,   # Transparency of the fill
           height=6,    # Figure height
           aspect=1.5)  # Aspect ratio

# Customize the plot
plt.title('Distribution of Top 5 Accept Rating Criteria by Name Category', 
          fontsize=14, 
          pad=15)
plt.xlabel('Top 5 Accept Rating Criteria (Mean)', fontsize=12)
plt.ylabel('Density', fontsize=12)

# Add legend with a better position
plt.legend(title='Name Category', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout to prevent label cutoff
plt.tight_layout()

# Show the plot
plt.show()

# Alternative: Create separate distribution plots for each category
plt.figure(figsize=(12, 6))

# Create individual distribution plots using FacetGrid
g = sns.FacetGrid(df, col='Name_Category', col_wrap=2, height=4)
g.map_dataframe(sns.kdeplot, x='top5_accept_rating_criteria_mean', fill=True)
g.fig.suptitle('Distribution of Top 5 Accept Rating Criteria by Name Category', 
               y=1.02, 
               fontsize=14)

# Add labels
g.set_axis_labels('Top 5 Accept Rating Criteria (Mean)', 'Density')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
