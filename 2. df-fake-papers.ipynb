```python
import pandas as pd
import numpy as np

fields = ["mic", "mac", "app", "fin"]
models = ["4o", "o1", "claude", "llama", "gemma", "deepseek"]
files = np.array([[[f"{model}-{field}_{i}" for i in range(5)] for field in fields] for model in models]).flatten().tolist()
```

```python
df = pd.DataFrame(columns=['id', 'file', 'name', 'journal', 'authors', 'affiliations',
       'len-original', 'len-anond'])

for file in files:
    with open(f'./dump/fake-papers/{file}.txt', 'r') as f:
        text = f.read()
        l = len(text.split())
        df.loc[len(df)] = [file, file+".txt", "", "", "", "", l, l]
```

```python
from tabulate import tabulate
print(tabulate(df, headers='keys', tablefmt='psql'))
```

```python
df.to_csv('dump/csv/fake_papers.csv', index=False)
```
